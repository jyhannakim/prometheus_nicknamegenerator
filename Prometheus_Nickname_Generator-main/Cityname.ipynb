# random 써서 키워드 추출

```python
import random
```

```python
# city 이름 들어가면 키워드 리턴하는 함수

def city_name(word):
  if word in city:
    if(len(city[word])==1):
      return city[word]
    else:
      value_list=city[word]
      return value_list[random.randint(0, len(value_list))]
```

# 테스트

```python
#random 안써도 되는거
city_name('대전')
```
